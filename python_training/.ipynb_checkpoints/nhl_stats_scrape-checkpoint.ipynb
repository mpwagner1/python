```python
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Chrome()
```

```python
driver.get( "https://www.nhl.com/stats/skaters?reportType=season&seasonFrom=20152016&seasonTo=20152016&gameType=2&filter=gamesPlayed,gte,1&sort=points,goals,assists&page=0&pageSize=100"
)
```

```python
html_content = driver.page_source
```

```python
driver.quit()
```

```python
soup = BeautifulSoup(html_content, "html.parser")
```

```python
print(soup)
```

```python
soup.find_all('div', role="columnheader")
```

```python
headers = soup.find_all('div', role="columnheader")
```

```python
headers_table = [headers.text for headers in headers]
```

```python
print (headers_table)
```

```python
import pandas as pd
```

```python
nhl_stats_df = pd.DataFrame(columns = headers_table)
nhl_stats_df = nhl_stats_df.drop(columns=["i"])
nhl_stats_df
```

```python
soup.find_all('div', class_="rt-td")
```

```python
columns = soup.find_all('div', class_="rt-td")
```

```python
rows = soup.find_all("div", role="row")
```

```python
player_data = []

for row in rows:
    row_data = row.find_all("div", class_="rt-td")
    individual_row_data = [data.text.strip() for data in row_data[1:]]
    formatted_row_data = ' '.join(individual_row_data)
    player_data.append(individual_row_data)
```

```python
player_data_df = pd.DataFrame(player_data)
```

```python
player_data_df = player_data_df.drop(player_data_df.index[0])
player_data_df
```

```python
updated_column_names = nhl_stats_df.columns
player_data_df.columns = updated_column_names
player_data_df
```
